```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

```python
text = open('대한민국헌법.txt', encoding='utf-8').read()
```

```python
text
```

```python
okt = Okt()
```

```python
nouns = okt.nouns(text)
```

```python
word = [n for n in nouns if len(n) > 1]
```

```python
word
```

```python
cnt = Counter(word)
cnt
```

```python
wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250, max_words=50)
gen = wc.generate_from_frequencies(cnt)
plt.figure()
plt.axis('off')
plt.imshow(gen)
```
